```python
import numpy as np
# import matplotlib.pyplot as plt

opt = 'milp_ol'
timelimit_gurobi = 240
N=40
n_workers=10
test_data_compressed = np.load('tests//tests_learning_%s_%d_%.3d_%.3d.npy' %(opt,N,timelimit_gurobi,0), allow_pickle=True).item()

array_minlp_nostop_ws_10m = test_data_compressed['array_minlp_nostop_ws_10m']
array_minlp_nostop = test_data_compressed['array_minlp_nostop']
array_minlp_nostop_ws = test_data_compressed['array_minlp_nostop_ws']
array_minlp = test_data_compressed['array_minlp']
array_minlp_ws = test_data_compressed['array_minlp_ws']
array_milp = test_data_compressed['array_milp']
array_learning_lp = test_data_compressed['array_milp_lp']
array_learning_nlp = test_data_compressed['array_milp_nlp']
cntr_infeas_milp_nlp = test_data_compressed['cntr_infeas_milp_nlp']
cntr_infeas_milp_lp = test_data_compressed['cntr_infeas_milp_lp']
info = test_data_compressed['info']

for job_idx in range(1, n_workers):
    test_data_compressed = np.load('tests//tests_learning_%s_%d_%.3d_%.3d.npy' %(opt,N, timelimit_gurobi,job_idx), allow_pickle=True).item()
    
    array_minlp_nostop_ws_10m_tmp = test_data_compressed['array_minlp_nostop_ws_10m']
    array_minlp_nostop_tmp = test_data_compressed['array_minlp_nostop']
    array_minlp_nostop_ws_tmp = test_data_compressed['array_minlp_nostop_ws']
    array_minlp_tmp = test_data_compressed['array_minlp']
    array_minlp_ws_tmp = test_data_compressed['array_minlp_ws']
    array_milp_tmp = test_data_compressed['array_milp']
    array_learning_lp_tmp = test_data_compressed['array_milp_lp']
    array_learning_nlp_tmp = test_data_compressed['array_milp_nlp']
    
    array_minlp_nostop_ws_10m = np.concatenate((array_minlp_nostop_ws_10m, array_minlp_nostop_ws_10m_tmp))
    array_minlp_nostop = np.concatenate((array_minlp_nostop, array_minlp_nostop_tmp))    
    array_minlp_nostop_ws = np.concatenate((array_minlp_nostop_ws, array_minlp_nostop_ws_tmp))    
    array_minlp = np.concatenate((array_minlp, array_minlp_tmp))    
    array_minlp_ws = np.concatenate((array_minlp_ws, array_minlp_ws_tmp))
    array_milp = np.concatenate((array_milp, array_milp_tmp))
    array_learning_lp = np.concatenate((array_learning_lp, array_learning_lp_tmp))
    array_learning_nlp = np.concatenate((array_learning_nlp, array_learning_nlp_tmp))   
    cntr_infeas_milp_nlp+= test_data_compressed['cntr_infeas_milp_nlp']
    cntr_infeas_milp_lp+= test_data_compressed['cntr_infeas_milp_lp']
    
num_datapoints = array_minlp.shape[0]
idx_feasible_points = (array_learning_nlp[:,0]!=False)*(array_learning_lp[:,0]!=False)

array_minlp_nostop_ws_10m = array_minlp_nostop_ws_10m[idx_feasible_points]
array_minlp_nostop = array_minlp_nostop[idx_feasible_points]
array_minlp_nostop_ws = array_minlp_nostop_ws[idx_feasible_points]
array_minlp = array_minlp[idx_feasible_points]
array_minlp_ws = array_minlp_ws[idx_feasible_points]
array_milp = array_milp[idx_feasible_points]
array_learning_lp = array_learning_lp[idx_feasible_points]
array_learning_nlp = array_learning_nlp[idx_feasible_points]
```
