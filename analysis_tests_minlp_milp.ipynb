```python
import numpy as np
import matplotlib.pyplot as plt

n_workers=10
test_data_compressed = np.load('tests//tests_minlp_milp_%d.npy' %(0), allow_pickle=True).item()
array_minlp_nostop = test_data_compressed['array_minlp_nostop']
array_minlp = test_data_compressed['array_minlp']
array_minlp_nlp = test_data_compressed['array_minlp_nlp']
array_minlp_lp = test_data_compressed['array_minlp_lp']
array_milp = test_data_compressed['array_milp']
array_milp_nlp = test_data_compressed['array_milp_nlp']
array_milp_lp = test_data_compressed['array_milp_lp']
info = test_data_compressed['info']
for job_idx in range(1, n_workers):
    test_data_compressed = np.load('tests//tests_minlp_milp_%d.npy' %(job_idx), allow_pickle=True).item()
    
    array_minlp_nostop_tmp = test_data_compressed['array_minlp_nostop']
    array_minlp_tmp = test_data_compressed['array_minlp']
    array_minlp_nlp_tmp = test_data_compressed['array_minlp_nlp']
    array_minlp_lp_tmp = test_data_compressed['array_minlp_lp']
    array_milp_tmp = test_data_compressed['array_milp']
    array_milp_nlp_tmp = test_data_compressed['array_milp_nlp']
    array_milp_lp_tmp = test_data_compressed['array_milp_lp']
    
    array_minlp_nostop = np.concatenate((array_minlp_nostop, array_minlp_nostop_tmp))
    array_minlp = np.concatenate((array_minlp, array_minlp_tmp))
    array_minlp_nlp = np.concatenate((array_minlp_nlp, array_minlp_nlp_tmp))
    array_minlp_lp = np.concatenate((array_minlp_lp, array_minlp_lp_tmp))
    array_milp = np.concatenate((array_milp, array_milp_tmp))
    array_milp_nlp = np.concatenate((array_milp_nlp, array_milp_nlp_tmp))
    array_milp_lp = np.concatenate((array_milp_lp, array_milp_lp_tmp))
```

```python
print(info)
```

```python
print('mean runtime minlp (no stop): \t\t %.2f ' %(np.mean(array_minlp_nostop[:,0])))
print('mean runtime minlp: \t\t\t %.2f' %(np.mean(array_minlp[:,0])))
print('mean runtime minlp + nlp: \t\t %.2f' %(np.mean(array_minlp_nlp[:,0])))
print('mean runtime minlp + lp: \t\t %.2f' %(np.mean(array_minlp_lp[:,0])))
print('mean runtime milp: \t\t\t %.2f' %(np.mean(array_milp[:,0])))
print('mean runtime milp + nlp: \t\t %.2f' %(np.mean(array_milp_nlp[:,0])))
print('mean runtime milp + lp: \t\t %.2f' %(np.mean(array_milp_lp[:,0])))
```
